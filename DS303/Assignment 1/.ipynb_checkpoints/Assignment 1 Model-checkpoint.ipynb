# Assignment 1
Goal: Given the dataset, **predict** whether a flight will be **delayed** or **on time**.

### Setup Data

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report

%matplotlib inline
pd.set_option('display.max_columns', None)
```

```python
data = pd.read_csv('dataset/FlightDelays.csv')
display(data.head())
```

### Get rid of duplicate/irrelevant features
- FL_NUM & TAIL_NUM have a lot of catagories and dont seem relevant to flight delays
- FL_DATE has already been represented in DAY_OF_MONTH since the month and year of data are fixed
- Hence we can drop these 3 columns from the start itself

```python
data = data.drop({'FL_NUM', 'TAIL_NUM', 'FL_DATE'}, axis='columns', errors='ignore')
data.head()
```

### 1. Data Visualisation
- Analyse feature datatypes and check for null values
- Plotting histograms(continuous data) & count plots(catagorical data)

```python
# Check for missing values and analyse datatypes
display(data.info(), data.describe())
```

```python
data[{'CRS_DEP_TIME', 'DEP_TIME'}].hist(bins=50, figsize=(20, 5))
plt.savefig("images/figure0.png")

fig, ax = plt.subplots(1, 2, gridspec_kw={'width_ratios': [1, 2]}, figsize=(20, 5))
# ax[1].tick_params(axis='x', rotation=90)
sns.countplot(data=data, x='CARRIER', ax=ax[0])
sns.countplot(data=data, x='DAY_OF_MONTH', ax=ax[1])
plt.savefig("images/figure1.png")

fig, ax = plt.subplots(1, 2, figsize=(20, 5))
sns.countplot(data=data, x='DAY_WEEK', ax=ax[0])
sns.countplot(data=data, x='DISTANCE', ax=ax[1])
plt.savefig("images/figure2.png")

fig, ax = plt.subplots(1, 4, gridspec_kw={'width_ratios': [3, 3, 2, 2]}, figsize=(20, 5))
sns.countplot(data=data, x='DEST', ax=ax[0])
sns.countplot(data=data, x='ORIGIN', ax=ax[1])
sns.countplot(data=data, x='Weather', ax=ax[2])
sns.countplot(data=data, x='Flight Status', ax=ax[3])
plt.savefig("images/figure3.png")
```

### 2. Data Preprocessing
- Seperate data into independant(X) and dependant(Y) features
- Handle null/missing values (Imputer)
- Handle catagorical features by one hot encoding (create dummy variables)
- Split into 60% train and 40% test to avoid overfitting
- Scale all features

```python
# Seperate dependant & independant features
X = data.drop('Flight Status', axis='columns')
y = data['Flight Status'].copy()
```

```python
# Create dummy variables for catagorical values
X = pd.get_dummies(X, drop_first=True)
X.head()
```

```python
# Split data into training & testing
print("Before splitting: shape(X) =", X.shape)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)
print("After  splitting: shape(X_train) =", X_train.shape, "and shape(X_test) =", 
      X_test.shape)
```

```python
# Scale both train and test data based on training set
scaler = StandardScaler()
scaler.fit(X_train, y_train)

XS_train = pd.DataFrame(scaler.transform(X_train), columns=X_train.columns,
                        index=X_train.index)
XS_test = pd.DataFrame(scaler.transform(X_test), columns=X_test.columns,
                        index=X_test.index)
XS_train.head()
```

## 3. Classifier without feature selection
- To test use of feature selection
- Use Logistic Regression
- Analyse coefficient

```python
# Create Logistic Regression Classifier
classifier = LogisticRegression(penalty='l2', tol=1e-6, max_iter=10000)
classifier.fit(XS_train, y_train)

# Make our predictions
y_pred = classifier.predict(XS_test)

# Analyse our predictions
print(classification_report(y_test, y_pred))
```

```python
# Analyse the coeffients of our model
print(f"We have total {classifier.coef_.shape[1]} features.")
display(classifier.coef_)
XS_train.head()
```
