# Assignment 1
Goal: Given the dataset, **predict** whether a flight will be **delayed** or **on time**.

### Setup Data

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score

%matplotlib inline
pd.set_option('display.max_columns', None)
```

```python
data = pd.read_csv('dataset/FlightDelays.csv')
display(data.head())
```

### Get rid of duplicate/irrelevant features
- FL_NUM & TAIL_NUM have a lot of catagories and dont seem relevant to flight delays
- FL_DATE has already been represented in DAY_OF_MONTH since the month and year of data are fixed
- Hence we can drop these 3 columns from the start itself

```python
data = data.drop({'FL_NUM', 'TAIL_NUM', 'FL_DATE'}, axis='columns', errors='ignore')
data.head()
```

### 1. Data Visualisation
- Analyse feature datatypes and check for null values
- Plotting histograms(continuous data) & count plots(catagorical data)

```python
# Check for missing values and analyse datatypes
display(data.info(), data.describe())
```

```python
# Heatmap to analyse for missing values
sns.heatmap(data.isnull(), yticklabels=False, cbar=False)

# Count Plots & Histograms
data[{'CRS_DEP_TIME', 'DEP_TIME'}].hist(bins=100, figsize=(20, 5))
plt.savefig("images/figure0.png")

fig, ax = plt.subplots(1, 2, gridspec_kw={'width_ratios': [1, 2]}, figsize=(20, 5))
# ax[1].tick_params(axis='x', rotation=90)
sns.countplot(data=data, x='CARRIER', ax=ax[0])
sns.countplot(data=data, x='DAY_OF_MONTH', ax=ax[1])
plt.savefig("images/figure1.png")

fig, ax = plt.subplots(1, 2, figsize=(20, 5))
sns.countplot(data=data, x='DAY_WEEK', ax=ax[0])
sns.countplot(data=data, x='DISTANCE', ax=ax[1])
plt.savefig("images/figure2.png")

fig, ax = plt.subplots(1, 4, gridspec_kw={'width_ratios': [3, 3, 2, 2]}, figsize=(20, 5))
sns.countplot(data=data, x='DEST', ax=ax[0])
sns.countplot(data=data, x='ORIGIN', ax=ax[1])
sns.countplot(data=data, x='Weather', ax=ax[2])
sns.countplot(data=data, x='Flight Status', ax=ax[3])
plt.savefig("images/figure3.png")
```

### 2. Data Preprocessing
- Seperate data into independant(X) and dependant(Y) features
- Handle null/missing values (Imputer)
- Handle catagorical features by one hot encoding (create dummy variables)
- Split into 60% train and 40% test to avoid overfitting
- Scale all features

```python
# Seperate dependant & independant features
X = data.drop('Flight Status', axis='columns')
y = data['Flight Status'].copy()
```

```python
# Create dummy variables for catagorical values
X = pd.get_dummies(X, drop_first=True)
display(X.head())
```

```python
# Split data into training & testing
print("Before splitting: shape(X) =", X.shape)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)
print("After  splitting: shape(X_train) =", X_train.shape, "and shape(X_test) =", 
      X_test.shape)
```

```python
# Scale both train and test data based on training set
scaler = StandardScaler()
scaler.fit(X_train)

XS_train = pd.DataFrame(scaler.transform(X_train), columns=X_train.columns,
                        index=X_train.index)
XS_test = pd.DataFrame(scaler.transform(X_test), columns=X_test.columns,
                        index=X_test.index)
XS_train.head()
```

## 3. Classifier without feature selection
- To test use of feature selection
- Use Logistic Regression
- Analyse coefficient

```python
# Create Logistic Regression Classifier
clf = LogisticRegression(penalty='l2', tol=1e-6, max_iter=10000)
clf.fit(XS_train, y_train.values.ravel())

# Make our predictions
y_pred = clf.predict(XS_test)

# Analyse our predictions
print(classification_report(y_test, y_pred))
```

```python
# Analyse the coeffients of our model
print(f"We have total {clf.coef_.shape[1]} features (including dummy variables).")
print(f"coefficients: {clf.coef_[0]}")
print(f"intercept: {clf.intercept_[0]}")
print(f"features: {XS_train.columns}")
XS_train.head()
```

### 4. Feature Selection
- Convert day of week to weekend or not
- Use DEP_TIME and CRS_DEP_TIME to analyse if a flight is already delayed or not. IE convert it into a boolean feature which is 1 if CRS_DEP_TIME + 15 < DEP_TIME, else 0
- Using sine cos cyclic encoding for DAY_OF_MONTH data did not improve accuracy, hence we can drop it
- We will use Count Encoding for CARRIER
- We will drop DISTANCE, DEST and ORIGIN as they don't give much of an improvement (due to small logistic regression coefficients)
- Plot Correlation matrix to analyse interdependence of features

```python
data2 = data.copy()
data2['WEEKEND'] = np.where(data['DAY_WEEK'] < 6, 0, 1)
data2['LATE_DEP'] = np.where(data['CRS_DEP_TIME'] + 15 < data['DEP_TIME'], 1, 0)

# Count Encodint for CARRIER
# count_map = data2['CARRIER'].value_counts().to_dict()
# data2['CARRIER'] = data2['CARRIER'].map(count_map)

data2 = data2.drop({'DISTANCE', 'CRS_DEP_TIME', 'DEP_TIME', 'DAY_WEEK', 'DAY_OF_MONTH','DEST',
                    'ORIGIN', 'CARRIER'}, axis='columns')
display(data2.head())
```

```python
# Correlation Matrix
# status = {'ontime': 1, 'delayed': 0}
# data_corr['Flight Status'] = [status[flight] for flight in data['Flight Status']]
data_corr = pd.get_dummies(data2, drop_first=True)
cm = data_corr.corr()
fig, ax = plt.subplots(figsize=(5, 5))
sns.heatmap(cm, xticklabels=cm.columns, yticklabels=cm.columns, annot=True)
cm['Flight Status_ontime'].sort_values(ascending=False)
```

### 5. Fitting a new model

```python
X2 = data2.drop('Flight Status', axis='columns')
X2 = pd.get_dummies(X2, drop_first=True)
y2 = data2['Flight Status'].copy()

X2_train, X2_test, y2_train, y2_test = train_test_split(X2, y2, test_size=0.4, 
                                                        random_state=42)
```

```python
# Scaling down our data
scaler = StandardScaler().fit(X2_train, y2_train)
XS2_train = pd.DataFrame(scaler.transform(X2_train), columns=X2_train.columns,
                         index=X2_train.index)
XS2_test = pd.DataFrame(scaler.transform(X2_test), columns=X2_test.columns,
                        index=X2_test.index)
```

```python
# Fitting a new logreg model
clf = LogisticRegression(penalty='l2', tol=1e-6, max_iter=10000)
clf.fit(XS2_train, y2_train)

# Make our predictions
y2_pred = clf.predict(XS2_test)

# Analyse our predictions
sns.heatmap(confusion_matrix(y2_test, y2_pred), annot=True, annot_kws={"size": 16})
plt.xlabel('Predicted Label', size=16), plt.ylabel('Truth Label', size=16)

print("Accuracy after Feature Selection:", 
      f"{round(accuracy_score(y2_test, y2_pred) * 100)}%\n")
print(classification_report(y2_test, y2_pred))
```

```python
flight = pd.DataFrame(data={'Weather': [0, 0, 0, 0, 1, 1, 1, 1], 
                            'WEEKEND': [0, 0, 1, 1, 0, 0, 1, 1], 
                            'LATE_DEP':[0, 1, 0, 1, 0, 1, 0, 1]})
display(flight.head(8))

# Label 1 means ontime
print(f"P(ontime flight): {clf.predict_proba(flight)[:,1]}")
```
