# K Nearest Neighbors

## Steps:

#### 1. Standardize the variables
Scale all the features to the same size, because distance between values can have surprising effects in KNN. 

```python
%matplotlib inline
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from sklearn import preprocessing, model_selection, neighbors, metrics
```

```python
data = pd.read_csv('dataset/Classified Data.csv')
data = data.drop('Unnamed: 0', axis=1)
display(data)

X = data.drop('TARGET CLASS', axis=1)
y = data['TARGET CLASS']
```

```python
# Standardize the variables
scaler = preprocessing.StandardScaler()
scaler.fit(X)
X_scaled = pd.DataFrame(data=scaler.transform(X), columns=X.columns)
display(X_scaled)

# Split train and test data from the same set:
X_train, X_test, y_train, y_test = model_selection.train_test_split(X_scaled, y, test_size=0.3)
```

```python
knn = neighbors.KNeighborsClassifier(n_neighbors=42)
knn.fit(X_train, y_train)

# Make our model predict
y_pred = knn.predict(X_test)

# Analyse our prediction accuracy and visualise with a Confusion Matrix
cm = metrics.confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, annot_kws={"size": 16})
plt.xlabel('Predicted Label', size=16), plt.ylabel('Truth Label', size=16)
print("Accuracy: ", metrics.accuracy_score(y_test, y_pred))
```

```python
n = 100
err = [None] * n
for i in range(0, n):
    knn = neighbors.KNeighborsClassifier(n_neighbors=i+1)
    knn.fit(X_train, y_train)
    y_pred_i = knn.predict(X_test)
    err[i] = 100*metrics.accuracy_score(y_test, y_pred_i)

# Plot K value vs accuracy for KNN
plt.figure(0, figsize=(24, 5))
plt.plot(range(1, n+1), err, marker='o', markerfacecolor='red', markersize=5)
plt.xlabel('K value', size=18), plt.ylabel('% Accuracy', size=18)
plt.xticks(range(1, 102, 5))
print('From the graph, best value of K seems to be: 42')
```
